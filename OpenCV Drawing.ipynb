```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
canvas = np.zeros((300,300,3), dtype = "uint8")
```

```python
%matplotlib inline
plt.imshow(canvas)
#cv2.imshow("Canvas", canvas)
```

```python
plt.imshow(cv2.line(canvas, (0,0), (300,300), (255,0,0), 4))
plt.imshow(cv2.rectangle(canvas, (0,0), (300,300), (255,0,0), 8))
```

```python
half = int(canvas.shape[0]/2)
plt.imshow(cv2.circle(canvas, (half, half), half, (255,0,0)))
```
