```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("/home/divyanshu/Desktop/CV Basics/image.jpg")
img_gray = cv2.imread("/home/divyanshu/Desktop/CV Basics/image.jpg", 0)
beach = cv2.imread("/home/divyanshu/Desktop/CV Basics/beach.jpg")
beach_gray = cv2.imread("/home/divyanshu/Desktop/CV Basics/beach.jpg", 0)
plt.imshow(img_gray, cmap = "gray")
```

```python
plt.imshow(beach_gray, cmap = "gray")
```

```python
#Grayscale Histogram
gray_hist = cv2.calcHist([img_gray], [0], None, [256], [0,255])
gray_hist_beach = cv2.calcHist([beach_gray], [0], None, [256], [0,255])
```

```python
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("Number of Pixels")
plt.plot(gray_hist)
plt.xlim([0, 255])
plt.show()
```

```python
#For the 'New Ring' image
b,g,r = cv2.split(img)
b_hist = cv2.calcHist([b], [0], None, [256], [0,256])
g_hist = cv2.calcHist([g], [0], None, [256], [0,256])
r_hist = cv2.calcHist([r], [0], None, [256], [0,256])
plt.plot(b_hist, color = 'b')
plt.plot(g_hist, color = 'g')
plt.plot(r_hist, color = 'r')
```

```python
#For the Beach image

b,g,r = cv2.split(beach)
b_hist = cv2.calcHist([b], [0], None, [256], [0,256])
g_hist = cv2.calcHist([g], [0], None, [256], [0,256])
r_hist = cv2.calcHist([r], [0], None, [256], [0,256])
plt.plot(b_hist, color = 'b')
plt.plot(g_hist, color = 'g')
plt.plot(r_hist, color = 'r')
```

```python
#For the Beach image
b,g,r = cv2.split(beach)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,3,1)
bg_hist = cv2.calcHist([b, g], [0,1], None, [32, 32], [0, 256 , 0, 256])
p = ax.imshow(bg_hist, interpolation = "nearest")
plt.colorbar(p)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,3,1)
rb_hist = cv2.calcHist([r, b], [0,1], None, [32, 32], [0, 256 , 0, 256])
p = ax.imshow(rb_hist, interpolation = "nearest")
plt.colorbar(p)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,3,1)
gr_hist = cv2.calcHist([g, r], [0,1], None, [32, 32], [0, 256 , 0, 256])
p = ax.imshow(gr_hist, interpolation = "nearest")
plt.colorbar(p)
```

```python
#Histogram Equalization
eq_img = cv2.equalizeHist(img_gray)
eq_img_hist = cv2.calcHist([eq_img], [0], None, [256], [0,256])
plt.imshow(eq_img, cmap = "gray")
```

```python
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("Number of Pixels")
plt.plot(eq_img_hist)
plt.xlim([0, 255])
plt.show()
```

```python
eq_img_beach = cv2.equalizeHist(beach_gray)
eq_img_hist_beach = cv2.calcHist([eq_img_beach], [0], None, [256], [0,256])
plt.imshow(eq_img_beach, cmap = "gray")
```
