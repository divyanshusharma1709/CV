```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/home/divyanshu/Desktop/CV Basics/image.jpg", 0)
```

```python
plt.imshow(img, cmap = "gray")
```

```python
canvas = np.zeros(img.shape[:2], dtype = "uint8")
circle = cv2.circle(canvas, (1280//2,720//2), 720//2, 255, -1)
```

```python
plt.imshow(canvas, cmap = "gray")
print(img.shape)
```

```python
#Two Images 'ANDed'
cv_and = cv2.bitwise_and(img,canvas)
plt.imshow(cv_and, cmap = "gray")
```

```python
#Two Images 'OR d'
cv_or = cv2.bitwise_or(img,canvas)
plt.imshow(cv_or, cmap = "gray")
```

```python
#Two Images 'XOR d'

cv_xor = cv2.bitwise_xor(img,canvas)
plt.imshow(cv_xor, cmap = "gray")
```

```python
#The Image Inverted (NOT Operation)
cv_not = cv2.bitwise_not(img)
plt.imshow(cv_not, cmap = "gray")
```

```python
#Masking
mask = np.zeros(img.shape[:2], dtype = "uint8")
cv_masked = cv2.bitwise_and(img, img, mask = canvas)
plt.imshow(cv_masked, cmap = "gray")
```

```python
#DIFFERENCE BETWEEN 'BITWISE_AND' AND 'MASKING'

#In bitwise_and, the color intensities are bitwise ANDed i.e., convert intensity integers to binary and AND them,
#resulting in a new intensity
#In Masking, put mask on image. If mask exists, let the color through, else 0 i.e. Black. Kind of like a filter
```
