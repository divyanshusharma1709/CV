```python
import cv2
import numpy as np
import imutils
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("/home/divyanshu/Desktop/CV Exp/image.jpg")
%matplotlib inline
plt.imshow(image)
```

```python
def translate(image, x, y):
    M = np.float32([[1, 0, x], [0, 1, y]])
    rows, cols, ch = image.shape
    dst = cv2.warpAffine(image,M,(cols, rows))
    return dst
```

```python
x = translate(image, 100, 100)
plt.imshow(x)
```

```python
#Rotation
def rotate(image, angle, center = None, scale = 1.0):
    (h, w) = image.shape[:2]
    if center is None:
        center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))
    return rotated
```

```python
rota = rotate(image, 25, None, 1)
plt.imshow(rota)
```

```python
#Scaling
def resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    dim = None
    (h, w) = image.shape[:2]
    if width is None and height is None:
        return image

    if width is None:
        r = height / float(h)
        dim = (int(w * r), height)

    else:
        r = width / float(w)
        dim = (width, int(h * r))

    resized = cv2.resize(image, dim, interpolation = inter)
```

```python
plt.imshow(resize(image, None, 1280))
```
